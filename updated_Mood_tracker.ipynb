# Period Mood and Fertility Tracker

```python

import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import datetime
```

```python

def plot_mood_and_fertility_tracker_with_centered_dates(start_date):
    # Days for each phase and the fertile window
    menstruation_days = 5
    follicular_days = 9
    ovulation_days = 3
    luteal_days = 11
    fertile_window_start = 9
    fertile_window_end = 14

    # Calculate end dates for each phase
    end_menstruation = start_date + datetime.timedelta(days=menstruation_days)
    end_follicular = end_menstruation + datetime.timedelta(days=follicular_days)
    end_ovulation = end_follicular + datetime.timedelta(days=ovulation_days)
    end_luteal = end_ovulation + datetime.timedelta(days=luteal_days)

    # Mood levels (arbitrary scale, where higher values represent better moods)
    menstruation_mood = 3
    follicular_mood = 7
    ovulation_mood = 6
    luteal_mood = 4

    dates = [start_date, end_menstruation, end_follicular, end_ovulation, end_luteal]
    moods = [menstruation_mood, follicular_mood, ovulation_mood, luteal_mood, menstruation_mood]  # Starting next cycle

    fig, ax = plt.subplots(figsize=(12, 6))

    ax.plot(dates, moods, marker='o', color='b', linestyle='-')
    ax.fill_between(dates, moods, color='lightblue')

    # Highlighting the fertile window
    fertile_start_date = start_date + datetime.timedelta(days=fertile_window_start)
    fertile_end_date = start_date + datetime.timedelta(days=fertile_window_end)
    ax.axvspan(fertile_start_date, fertile_end_date, alpha=0.3, color='green', label='Fertile Window')
    
    # Annotating the fertile window dates in the center
    fertile_center_date = fertile_start_date + (fertile_end_date - fertile_start_date) / 2
    ax.text(fertile_center_date, 2, f'{fertile_start_date.strftime("%b %d")} - {fertile_end_date.strftime("%b %d")}', 
            ha='center', color='black', fontsize=12)

    # Setting labels and title
    ax.set_title('Period Mood and Fertility Tracker', fontsize=16)
    ax.set_ylabel('Mood Level', fontsize=14)
    ax.set_xlabel('Date', fontsize=14)
    ax.set_ylim(0, 10)
    ax.set_yticks(range(11))
    ax.set_yticklabels(['', 'Very Low', '', 'Low', '', 'Neutral', '', 'Good', '', 'Very Good', ''])
    ax.grid(True, which='both', linestyle='--', linewidth=0.5)
    ax.axhline(0, color='black', linewidth=0.7)
    ax.axvline(start_date, color='r', linestyle='--', linewidth=1)
    ax.axvline(end_menstruation, color='r', linestyle='--', linewidth=1)
    ax.axvline(end_follicular, color='r', linestyle='--', linewidth=1)
    ax.axvline(end_ovulation, color='r', linestyle='--', linewidth=1)

    # Formatting x-axis
    ax.xaxis.set_major_locator(mdates.DayLocator(interval=1))
    ax.xaxis.set_major_formatter(mdates.DateFormatter('%b %d'))
    plt.xticks(rotation=45)

    # Legends
    ax.legend(loc='upper left')

    # Phase annotations
    ax.text(start_date + datetime.timedelta(days=2), 1.5, 'Menstruation', color='black', fontsize=12, ha='center')
    ax.text(end_menstruation + datetime.timedelta(days=4), 8, 'Follicular Phase', color='black', fontsize=12, ha='center')
    ax.text(end_follicular + datetime.timedelta(days=1), 6.5, 'Ovulation', color='black', fontsize=12, ha='center')
    ax.text(end_ovulation + datetime.timedelta(days=5), 4.5, 'Luteal Phase', color='black', fontsize=12, ha='center')

    plt.tight_layout()
    plt.show()
```

## Input Date and Generate Plot

```python

# Input the start date of the period
year = int(input("Enter the year (e.g. 2023): "))
month = int(input("Enter the month (1-12): "))
day = int(input("Enter the day (1-31): "))

start_date = datetime.date(year, month, day)
plot_mood_and_fertility_tracker_with_centered_dates(start_date)
```
